```python
import torch
import torch.nn as nn

class SketchClassifier(nn.Module):
    def __init__(self, eva_model):
        """
        Initialize the SketchClassifier with a pre-trained EVA model.
        
        Args:
            eva_model (nn.Module): The original EVA model
        """
        super(SketchClassifier, self).__init__()
        
        # Store the key components of the EVA model
        self.patch_embed = eva_model.patch_embed
        self.cls_token = getattr(eva_model, 'cls_token', None)
        self.pos_embed = getattr(eva_model, 'pos_embed', None)
        self.pos_drop = eva_model.pos_drop
        self.blocks = eva_model.blocks
        self.norm = eva_model.norm
        self.fc_norm = eva_model.fc_norm
        
    def extract_features(self, x):
        """
        Extract features from the input before the classification head.
        
        Args:
            x (torch.Tensor): Input tensor of shape (B, C, H, W)
        
        Returns:
            torch.Tensor: Feature vector before classification
        """
        # Patch embedding
        x = self.patch_embed(x)
        
        # Add CLS token if exists
        if self.cls_token is not None:
            cls_tokens = self.cls_token.expand(x.shape[0], -1, -1)
            x = torch.cat((cls_tokens, x), dim=1)
        
        # Add positional embedding if exists
        if self.pos_embed is not None:
            x = x + self.pos_embed
        
        # Positional dropout
        x = self.pos_drop(x)
        
        # Pass through transformer blocks
        for blk in self.blocks:
            x = blk(x)
        
        # Apply norm if needed
        if self.norm is not None:
            x = self.norm(x)
        
        # Apply FC norm
        x = self.fc_norm(x)
        
        #  # Compute attention weights
        # attention_weights = torch.softmax(x.mean(dim=-1), dim=1)
        
        # # Weighted average pooling
        # features = (x * attention_weights.unsqueeze(-1)).sum(dim=1)
        features = x.mean(dim=1)
        
        return features
        
  
```

```python
# 3. Function to extract features
def extract_and_store_features(model, dataloader, device):
    model.eval()
    features_list = []
    labels_list = []
    
    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc="Extracting features"):
            images = images.to(device)
            features = model.extract_features(images)
            
            features_list.append(features.cpu())
            labels_list.append(labels)
    
    # Concatenate all features and labels
    features = torch.cat(features_list, dim=0)
    labels = torch.cat(labels_list, dim=0)
    
    return features, labels
```

```python
import torch
import h5py
import numpy as np
from tqdm import tqdm

def extract_and_store_features_chunked(model, dataloader, output_file, chunk_size=100):
    model.eval()
    
    # Calculate total size
    total_samples = len(dataloader.dataset)
    feature_dim = 1408  # EVA CLIP feature dimension
    
    # Create H5 file
    with h5py.File(output_file, 'w') as f:  # 'w' mode to create new file
        # Create datasets
        features_dset = f.create_dataset('features', 
                                      shape=(total_samples, feature_dim),
                                      dtype='float32')
        labels_dset = f.create_dataset('labels', 
                                     shape=(total_samples,),
                                     dtype='int64')

        # Process in batches
        start_idx = 0
        with torch.no_grad():
            for images, labels in tqdm(dataloader, desc=f"Extracting features for {output_file}"):
                # Move to GPU and extract features
                images = images.to(device)
                features = model.extract_features(images)
                
                # Move back to CPU and convert to numpy
                batch_size = features.size(0)
                features = features.cpu().numpy()
                labels = labels.numpy()
                
                # Store in H5 file
                end_idx = start_idx + batch_size
                features_dset[start_idx:end_idx] = features
                labels_dset[start_idx:end_idx] = labels
                
                start_idx = end_idx
                
                # Clear cache
                torch.cuda.empty_cache()

```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import h5py
from tqdm import tqdm
import numpy as np
```

```python
# Custom Dataset for H5 features
class H5FeatureDataset(Dataset):
    def __init__(self, h5_path):
        self.h5_path = h5_path
        # Get dataset length
        with h5py.File(h5_path, 'r') as f:
            self.length = len(f['labels'])
    
    def __getitem__(self, idx):
        # Load single item from H5
        with h5py.File(self.h5_path, 'r') as f:
            feature = torch.from_numpy(f['features'][idx]).float()
            label = torch.tensor(f['labels'][idx]).long()
        return feature, label
    
    def __len__(self):
        return self.length
```

```python
# Classifier Model
class SketchClassifier(nn.Module):
    def __init__(self, input_dim=1408, hidden_dim=512, num_classes=10):
        super().__init__()
        self.classifier = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, num_classes)
        )
    
    def forward(self, x):
        return self.classifier(x)
```

```python
# Training setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create datasets and dataloaders
train_dataset = H5FeatureDataset('/kaggle/input/data-no-aug/train_features (1).h5')
val_dataset = H5FeatureDataset('/kaggle/input/data-no-aug/val_features (1).h5')

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=4)
```

```python

num_classes=500
model = SketchClassifier(num_classes=num_classes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=1e-3, weight_decay=0.01)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=30)
```

```python
# Training function
def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    pbar = tqdm(loader, desc='Training')
    for features, labels in pbar:
        features, labels = features.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(features)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()
        
        pbar.set_postfix({'loss': running_loss/total, 'acc': 100.*correct/total})
    
    return running_loss/len(loader), 100.*correct/total

# Validation function
@torch.no_grad()
def validate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for features, labels in tqdm(loader, desc='Validating'):
        features, labels = features.to(device), labels.to(device)
        outputs = model(features)
        loss = criterion(outputs, labels)
        
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()
    
    return running_loss/len(loader), 100.*correct/total

# Training loop
num_epochs = 30
best_val_acc = 0

print("Starting training...")
for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")
    
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    
    scheduler.step()
    
    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
    print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
    
    # Save best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'best_val_acc': best_val_acc,
        }, 'best_sketch_classifier.pt')
        print(f"New best model saved! Validation Accuracy: {val_acc:.2f}%")

print("\nTraining finished!")
print(f"Best validation accuracy: {best_val_acc:.2f}%")
```

# Scalable Contrastive Learning

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math

class LargeScaleContrastiveNet(nn.Module):
    def __init__(self, input_dim=1408, hidden_dim=2048, projection_dim=768, num_classes=500):
        super().__init__()
        
        # Larger encoder for better feature discrimination
        self.encoder = nn.Sequential(
            nn.LayerNorm(input_dim),
            nn.Linear(input_dim, hidden_dim),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2)
        )
        
        # Projection head with larger dimensions
        self.projection = nn.Sequential(
            nn.Linear(hidden_dim // 2, hidden_dim // 2),
            nn.GELU(),
            nn.Linear(hidden_dim // 2, projection_dim)
        )
        
        # Classification head with progressive reduction
        self.classifier = nn.Sequential(
            nn.Linear(hidden_dim // 2, hidden_dim // 2),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim // 2, hidden_dim // 4),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim // 4, num_classes)
        )
        
        # Initialize weights
        self.apply(self._init_weights)
    
    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            # Use Kaiming initialization for better gradient flow
            nn.init.kaiming_normal_(module.weight, mode='fan_out', nonlinearity='relu')
            if module.bias is not None:
                nn.init.zeros_(module.bias)
    
    def forward(self, x):
        features = self.encoder(x)
        projections = F.normalize(self.projection(features), dim=1)
        logits = self.classifier(features)
        return projections, logits

class MultiClassSupConLoss(nn.Module):
    """
    Supervised Contrastive Learning loss optimized for many classes
    """
    def __init__(self, temperature=0.1, base_temperature=0.07):
        super().__init__()
        self.temperature = temperature
        self.base_temperature = base_temperature
    
    def forward(self, features, labels, mask=None):
        device = features.device
        batch_size = features.shape[0]
        
        # Compute similarity matrix
        sim_matrix = torch.mm(features, features.T)
        
        # Loss computational optimization for large number of classes
        sim_matrix = sim_matrix / self.temperature
        sim_matrix_max, _ = torch.max(sim_matrix, dim=1, keepdim=True)
        sim_matrix = sim_matrix - sim_matrix_max.detach()  # For numerical stability
        
        # Create label mask
        labels = labels.contiguous().view(-1, 1)
        mask = torch.eq(labels, labels.T).float().to(device)
        
        # Remove self-contrast
        logits_mask = torch.scatter(
            torch.ones_like(mask),
            1,
            torch.arange(batch_size).view(-1, 1).to(device),
            0
        )
        mask = mask * logits_mask
        
        # Compute log probability
        exp_sim_matrix = torch.exp(sim_matrix) * logits_mask
        log_prob = sim_matrix - torch.log(exp_sim_matrix.sum(1, keepdim=True))
        
        # Compute mean of log-likelihood over positive pairs
        mean_log_prob_pos = (mask * log_prob).sum(1) / (mask.sum(1) + 1e-8)
        
        # Loss
        loss = -(self.temperature / self.base_temperature) * mean_log_prob_pos
        return loss.mean()

def train_epoch_large_scale(model, train_loader, optimizer, epoch, 
                          temperature=0.1, lambda_contrast=0.5,
                          gradient_accumulation_steps=2):
    """
    Training loop optimized for large number of classes
    """
    model.train()
    total_loss = 0
    total_con_loss = 0
    total_ce_loss = 0
    correct = 0
    total = 0
    
    criterion_ce = nn.CrossEntropyLoss()
    criterion_con = MultiClassSupConLoss(temperature=temperature)
    
    optimizer.zero_grad()
    
    for batch_idx, (features, labels) in enumerate(train_loader):
        features = features.cuda()
        labels = labels.cuda()
        
        # Forward pass
        projections, logits = model(features)
        
        # Compute losses
        con_loss = criterion_con(projections, labels)
        ce_loss = criterion_ce(logits, labels)
        
        # Combined loss
        loss = lambda_contrast * con_loss + (1 - lambda_contrast) * ce_loss
        
        # Gradient accumulation for larger effective batch size
        loss = loss / gradient_accumulation_steps
        loss.backward()
        
        if (batch_idx + 1) % gradient_accumulation_steps == 0:
            optimizer.step()
            optimizer.zero_grad()
        
        # Compute accuracy
        with torch.no_grad():
            _, predicted = logits.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
        
        total_loss += loss.item() * gradient_accumulation_steps
        total_con_loss += con_loss.item()
        total_ce_loss += ce_loss.item()
    
    avg_loss = total_loss / len(train_loader)
    avg_con_loss = total_con_loss / len(train_loader)
    avg_ce_loss = total_ce_loss / len(train_loader)
    accuracy = 100. * correct / total
    
    print(f'Epoch {epoch} Training: Loss: {avg_loss:.3f}, '
          f'Con Loss: {avg_con_loss:.3f}, CE Loss: {avg_ce_loss:.3f}, '
          f'Acc: {accuracy:.2f}%')
    
    return avg_loss, accuracy

def validate(model, val_loader, temperature=0.1, lambda_contrast=0.5):
    """
    Validation function
    """
    model.eval()
    total_loss = 0
    total_con_loss = 0
    total_ce_loss = 0
    correct = 0
    total = 0
    
    criterion_ce = nn.CrossEntropyLoss()
    criterion_con = MultiClassSupConLoss(temperature=temperature)
    
    with torch.no_grad():
        for features, labels in val_loader:
            features = features.cuda()
            labels = labels.cuda()
            
            projections, logits = model(features)
            
            con_loss = criterion_con(projections, labels)
            ce_loss = criterion_ce(logits, labels)
            loss = lambda_contrast * con_loss + (1 - lambda_contrast) * ce_loss
            
            _, predicted = logits.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
            
            total_loss += loss.item()
            total_con_loss += con_loss.item()
            total_ce_loss += ce_loss.item()
    
    avg_loss = total_loss / len(val_loader)
    avg_con_loss = total_con_loss / len(val_loader)
    avg_ce_loss = total_ce_loss / len(val_loader)
    accuracy = 100. * correct / total
    
    print(f'Validation: Loss: {avg_loss:.3f}, '
          f'Con Loss: {avg_con_loss:.3f}, CE Loss: {avg_ce_loss:.3f}, '
          f'Acc: {accuracy:.2f}%')
    
    return avg_loss, accuracy

def create_training_components(input_dim=1408, num_classes=500):
    """
    Initialize model and training components
    """
    model = LargeScaleContrastiveNet(
        input_dim=input_dim,
        hidden_dim=2048,
        projection_dim=768,
        num_classes=num_classes
    ).cuda()
    
    # Use AdamW with larger batch and learning rate warmup
    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=1e-4,
        weight_decay=0.05,
        betas=(0.9, 0.999)
    )
    
    # Cosine scheduler with warmup
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=3e-4,
        epochs=30,
        steps_per_epoch=len(train_loader),
        pct_start=0.1,
        anneal_strategy='cos'
    )
    
    return model, optimizer, scheduler
```

```python
def validate_top5(model, val_loader, temperature=0.1, lambda_contrast=0.5):
    """
    Validation function with Top-5 accuracy computation.
    """
    model.eval()
    total_loss = 0
    total_con_loss = 0
    total_ce_loss = 0
    correct_top1 = 0
    correct_top5 = 0
    total = 0
    
    criterion_ce = nn.CrossEntropyLoss()
    criterion_con = MultiClassSupConLoss(temperature=temperature)
    
    with torch.no_grad():
        for features, labels in val_loader:
            features = features.cuda()
            labels = labels.cuda()
            
            projections, logits = model(features)
            
            con_loss = criterion_con(projections, labels)
            ce_loss = criterion_ce(logits, labels)
            loss = lambda_contrast * con_loss + (1 - lambda_contrast) * ce_loss
            
            # Top-1 Accuracy
            _, predicted = logits.max(1)
            correct_top1 += predicted.eq(labels).sum().item()
            
            # Top-5 Accuracy
            top5_preds = torch.topk(logits, k=5, dim=1).indices
            correct_top5 += sum([labels[i] in top5_preds[i] for i in range(labels.size(0))])
            
            total += labels.size(0)
            
            total_loss += loss.item()
            total_con_loss += con_loss.item()
            total_ce_loss += ce_loss.item()
    
    avg_loss = total_loss / len(val_loader)
    avg_con_loss = total_con_loss / len(val_loader)
    avg_ce_loss = total_ce_loss / len(val_loader)
    top1_accuracy = 100. * correct_top1 / total
    top5_accuracy = 100. * correct_top5 / total
    
    print(f'Validation: Loss: {avg_loss:.3f}, '
          f'Con Loss: {avg_con_loss:.3f}, CE Loss: {avg_ce_loss:.3f}, '
          f'Top-1 Acc: {top1_accuracy:.2f}%, Top-5 Acc: {top5_accuracy:.2f}%')
    
    return avg_loss, top1_accuracy, top5_accuracy
```

```python


class EVA_finetuner(nn.Module):
    def __init__(self, model, num_classes=500):
        """
        Initializes the EVA model, modifying only the classifier layer to match the number of classes.

        Parameters:
        - model: The pre-trained Eva model.
        - num_classes: The number of classes in the new dataset.
        """
        super(EVA_finetuner, self).__init__()

        # Keep the rest of the model as it is
        self.model = model

        # Freeze all layers except the classifier (head)
        for param in self.model.parameters():
            param.requires_grad = True

        # Modify the final classifier layer to match the number of classes
        self.model.head = nn.Linear(in_features=1408, out_features=num_classes)

        # Ensure that the new classifier layers will be trained
        for param in self.model.head.parameters():
            param.requires_grad = True

    def forward(self, x):
        """
        Forward pass through the modified model.
        """
        return self.model(x)
```
