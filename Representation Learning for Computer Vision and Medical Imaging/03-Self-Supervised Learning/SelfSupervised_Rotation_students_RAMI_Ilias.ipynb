Now it's time to implement our network.

Since STL10 comes from ImageNet, we take the same network proposed for ImageNet in [1], inspired by AlexNet.

Here you can find the (modified) version of AlexNet (e.g., no DropOut) borrowed from the GitHub of the authors.  

**Question**: Choose an approriate value for num_classes and explain why



> The appropriate value for num_classes is 4, since the model is designed to predict which of the four geometric rotations (0°, 90°, 180°, 270°) was applied to the input image. This makes the rotation prediction task a 4-class classification problem.


```python
# Code borrowed from https://github.com/gidariss/FeatureLearningRotNet
# AlexNet
class AlexNet(nn.Module):
    def __init__(self):
        super(AlexNet, self).__init__()
        num_classes = 4

        conv1 = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=11, stride=4, padding=2),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
        )
        pool1 = nn.MaxPool2d(kernel_size=3, stride=2)
        conv2 = nn.Sequential(
            nn.Conv2d(64, 192, kernel_size=5, padding=2),
            nn.BatchNorm2d(192),
            nn.ReLU(inplace=True),
        )
        pool2 = nn.MaxPool2d(kernel_size=3, stride=2)
        conv3 = nn.Sequential(
            nn.Conv2d(192, 384, kernel_size=3, padding=1),
            nn.BatchNorm2d(384),
            nn.ReLU(inplace=True),
        )
        conv4 = nn.Sequential(
            nn.Conv2d(384, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
        )
        conv5 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
        )
        pool5 = nn.MaxPool2d(kernel_size=3, stride=2)

        num_pool5_feats = 6 * 6 * 256
        fc_block = nn.Sequential(
            nn.Flatten(),
            nn.Linear(num_pool5_feats, 4096, bias=False),
            nn.BatchNorm1d(4096),
            nn.ReLU(inplace=True),
            nn.Linear(4096, 4096, bias=False),
            nn.BatchNorm1d(4096),
            nn.ReLU(inplace=True),
        )
        classifier = nn.Sequential(
            nn.Linear(4096, num_classes),
        )

        self._feature_blocks = nn.ModuleList([
            conv1,
            pool1,
            conv2,
            pool2,
            conv3,
            conv4,
            conv5,
            pool5,
            fc_block,
            classifier,
        ])
        self.all_feat_names = [
            'conv1',
            'pool1',
            'conv2',
            'pool2',
            'conv3',
            'conv4',
            'conv5',
            'pool5',
            'fc_block',
            'classifier',
        ]
        assert(len(self.all_feat_names) == len(self._feature_blocks))

    def _parse_out_keys_arg(self, out_feat_keys):

        # By default return the features of the last layer / module.
        out_feat_keys = [self.all_feat_names[-1],] if out_feat_keys is None else out_feat_keys

        if len(out_feat_keys) == 0:
            raise ValueError('Empty list of output feature keys.')
        for f, key in enumerate(out_feat_keys):
            if key not in self.all_feat_names:
                raise ValueError('Feature with name {0} does not exist. Existing features: {1}.'.format(key, self.all_feat_names))
            elif key in out_feat_keys[:f]:
                raise ValueError('Duplicate output feature key: {0}.'.format(key))

        # Find the highest output feature in `out_feat_keys
        max_out_feat = max([self.all_feat_names.index(key) for key in out_feat_keys])

        return out_feat_keys, max_out_feat

    def forward(self, x, out_feat_keys=None):
        """Forward an image `x` through the network and return the asked output features.
        Args:
          x: input image.
          out_feat_keys: a list/tuple with the feature names of the features
                that the function should return. By default the last feature of
                the network is returned.
        Return:
            out_feats: If multiple output features were asked then `out_feats`
                is a list with the asked output features placed in the same
                order as in `out_feat_keys`. If a single output feature was
                asked then `out_feats` is that output feature (and not a list).
        """
        out_feat_keys, max_out_feat = self._parse_out_keys_arg(out_feat_keys)
        out_feats = [None] * len(out_feat_keys)

        feat = x
        for f in range(max_out_feat+1):
            feat = self._feature_blocks[f](feat)
            key = self.all_feat_names[f]
            if key in out_feat_keys:
                out_feats[out_feat_keys.index(key)] = feat

        out_feats = out_feats[0] if len(out_feats)==1 else out_feats
        return out_feats

    def get_L1filters(self):
        convlayer = self._feature_blocks[0][0]
        batchnorm = self._feature_blocks[0][1]
        filters = convlayer.weight.data
        scalars = (batchnorm.weight.data / torch.sqrt(batchnorm.running_var + 1e-05))
        filters = (filters * scalars.view(-1, 1, 1, 1).expand_as(filters)).cpu().clone()

        return filters
```

We build the RotationPrediction Module.

As in [1], we use the 'conv5' layer as representation features and a small classifier, as proposed in the paper.

The key part is in the 'forward' function and in '_preprocess'.

**Question**: complete the code XXXXXX in the '_preprocess' function. Remember that we are implementing the method of [1] as shown in the figure at the beginning of this TP.

```python
## Joint optimization
class RotationPrediction(nn.Module):
    metrics = ['Loss', 'Acc1']
    metrics_fmt = [':.4e', ':6.2f']

    def __init__(self, n_classes_classifier, device):
        super().__init__()
        self.device=device
        self.model = AlexNet().to(self.device)
        self.latent_dim = 256 * 6 * 6 # as defined in [1]
        self.feat_layer = 'conv5' # as defined in [1]
        self.n_classes_classifier = n_classes_classifier

    # Same Classifier as defined in [1]
    def construct_classifier(self):
        classifier = nn.Sequential(
            nn.AdaptiveMaxPool2d((6, 6)),
            nn.BatchNorm2d(256, affine=False),
            nn.Flatten(),
            nn.Linear(self.latent_dim, self.n_classes_classifier)
        )
        classifier=classifier.to(self.device)
        return classifier

    def forward(self, images):
        images=images.to(self.device)
        batch_size = images.shape[0]

        #images must have a torch.Size([batch_size, Channels, Height, Width])
        #target must have a torch.Size([batch_size])
        images, targets = self._preprocess(images)
        targets = targets.to(self.device)

        logits, zs = self.model(images, out_feat_keys=('classifier', self.feat_layer))
        loss = F.cross_entropy(logits, targets).to(self.device)

        pred = logits.argmax(dim=-1)
        correct = pred.eq(targets).float().sum()
        acc = correct / targets.shape[0] * 100.

        zs = zs[:batch_size]
        return dict(Loss=loss, Acc1=acc), zs[:batch_size]

    def encode(self, images, flatten=True):
        zs = self.model(images, out_feat_keys=(self.feat_layer,))
        return zs.flatten(start_dim=1)

    def _preprocess(self, images):
        batch_size = images.shape[0]
        # Create rotated versions of the images using flip and transpose operations
        images_0 = images  # 0° rotation (original image)
        images_90 = torch.flip(images.transpose(-2, -1), dims=(-2,))  # 90° rotation
        images_180 = torch.flip(torch.flip(images, dims=(-2,)), dims=(-1,))  # 180° rotation
        images_270 = torch.flip(images.transpose(-2, -1), dims=(-1,))  # 270° rotation

        # Concatenate all rotated images into a single batch
        images_batch = torch.cat([images_0, images_90, images_180, images_270], dim=0)

        # Create labels: [0,0,...,1,1,...,2,2,...,3,3,...] (each repeated batch_size times)
        targets = torch.arange(4, device=self.device).repeat_interleave(batch_size)

        return images_batch, targets
```

Here we train our algorithm and evaluate its prediction power on a test set (function 'validate') using the function `accuracy`.

**Question**: Do you understand what the function accuracy compute ? Explain what the output `res` contains



> The accuracy function computes the top-k accuracy of the model's predictions compared to the ground truth labels. It works by first determining the top k predicted classes for each input using output.topk, then comparing these predictions to the true labels (target) to count how many correct predictions fall within the top k. The function returns a list of accuracies (res) for each value of k specified in topk, expressed as percentages. For example, if topk=(1, 5), it computes both top-1 accuracy (the true label is the top prediction) and top-5 accuracy (the true label is among the top 5 predictions). The results are scaled to percentages by multiplying by 100.0 / batch_size.



```python
def accuracy(output, target, topk=(1,)):

    with torch.no_grad():
        maxk = max(topk)
        batch_size = target.size(0)

        _, pred = output.topk(k=maxk, dim=1, largest=True, sorted=True) # sliced torch.argsort (for each row, it outputs the indices containing the k max values)
        pred = pred.t() # transpose to get batches as last dimension
        correct = pred.eq(target.view(1, -1).expand_as(pred)) # first transpose then expand target to be the same size as pred and then elementwise-equality

        # compute number of samples in the batch that has correct label in the k highest predictions
        res = []
        for k in topk:
            correct_k = correct[:k].reshape(-1).float().sum(0, keepdim=True)
            res.append(correct_k.mul_(100.0 / batch_size))
        return res
```

Now we can define our train function.


**Question**: Look at the optimization process and describe it. How do we minimize the rotation prediction module and classifier ? Could you do it differently ?



> The optimization process minimizes the rotation prediction module (self-supervised task) and the classifier (supervised task) separately. First, the model predicts rotations for input images, and the rotation prediction loss (out['Loss']) is backpropagated to update the model's parameters using optimizer. The features (zs) extracted by the model are detached to prevent gradients from the classifier from affecting the model. Next, the classifier uses these detached features to predict class labels, and the cross-entropy loss is backpropagated to update only the classifier's parameters using optimizer_classifier. This decoupled approach ensures the self-supervised task drives feature learning without interference from the supervised task. Alternatively, joint optimization or weighted combined loss could be used, but the current method effectively balances both tasks.



```python
def train(train_loader, model, classifier, optimizer, optimizer_classifier, epoch, device):

    model.train()
    classifier.train()
    top1=[]
    top5=[]

    for i, (images, target) in enumerate(train_loader):
        # compute loss
        bs = images.shape[0]
        images = images.to(device)
        target = target.to(device)
        out, zs = model(images)
        zs = zs.detach() # detach from the graph,  requires_grad = False

        # compute optimizer step for ssl task using the previously defined Loss function of the model
        optimizer.zero_grad()
        out['Loss'].backward()
        optimizer.step()

        # compute loss
        logits = classifier(zs)
        loss = F.cross_entropy(logits, target)

        # compute accuracy
        acc1, acc5 = accuracy(logits, target, topk=(1, 5))
        top1.append(acc1[0])
        top5.append(acc5[0])

        # computeoptimizer step for classifier using the cross entropy loss
        optimizer_classifier.zero_grad()
        loss.backward()
        optimizer_classifier.step()

        #if i % 20 == 0:
        #    print('Epoch {}, Iter: {} out of {}, Average loss: {:.4f}, acc 1: {:.4f}'.format(epoch, i, len(train_loader), loss / len(train_loader.dataset),acc1[0]))

    print('Epoch: {}, Average loss: {:.4f}, Average acc 1: {:.4f}, Average acc 5: {:.4f}'.format(epoch, loss / len(train_loader.dataset),sum(top1)/len(top1), sum(top5)/len(top5)))
    return sum(top1)/len(top1), sum(top5)/len(top5)
```

**Question**: complete the code to compute average loss and average acc 1 and average acc 5 on the test set

```python
def test(test_loader, model, classifier, device):
    model.eval()
    classifier.eval()

    total_loss = 0.0
    total_samples = 0
    top1 = []
    top5 = []

    with torch.no_grad():
        for i, (images, target) in enumerate(test_loader):
            images = images.to(device)
            target = target.to(device)

            _, features = model(images)
            features = features.detach()
            logits = classifier(features)
            loss = F.cross_entropy(logits, target)

            batch_size = images.size(0)
            total_loss += loss.item() * batch_size
            total_samples += batch_size

            acc1, acc5 = accuracy(logits, target, topk=(1, 5))
            top1.append(acc1.item())
            top5.append(acc5.item())

    average_loss = total_loss / total_samples
    average_acc1 = sum(top1) / len(top1)
    average_acc5 = sum(top5) / len(top5)

    print(f'Test Set - Average Loss: {average_loss:.4f}, Top-1 Accuracy: {average_acc1:.2f}%, Top-5 Accuracy: {average_acc5:.2f}%')



    return average_loss, average_acc1, average_acc5
```

```python
# First stage: Train rotation prediction model
import torch
import torch.nn.functional as F

def train_rotation_model(train_loader, model, optimizer, epochs, device):
    rotation_losses = []
    rotation_accuracies = []

    best_loss = float('inf')  # Initialize best loss to infinity

    for epoch in range(epochs):
        model.train()
        epoch_losses = []
        epoch_accuracies = []

        for images, _ in train_loader:
            images = images.to(device)
            # Only compute rotation prediction loss and accuracy
            metrics, _ = model(images)
            loss = metrics['Loss']
            accuracy = metrics['Acc1']

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            epoch_losses.append(loss.item())
            epoch_accuracies.append(accuracy.item())

        avg_loss = sum(epoch_losses) / len(epoch_losses)
        avg_acc = sum(epoch_accuracies) / len(epoch_accuracies)
        print(f'Rotation Training - Epoch {epoch}: Loss = {avg_loss:.4f}, Accuracy = {avg_acc:.2f}%')

        rotation_losses.append(avg_loss)
        rotation_accuracies.append(avg_acc)

        # Save the model checkpoint if the current epoch's loss is lower than the best seen so far
        if avg_loss < best_loss:
            best_loss = avg_loss
            torch.save(model.state_dict(), "rotation_model_best.pth")
            print(f"Improved loss detected. Saving rotation model checkpoint at epoch {epoch}.")

    return rotation_losses, rotation_accuracies

def train_classifier(train_loader, model, classifier, optimizer_classifier, epochs, device):
    model.eval()  # Put feature extractor in eval mode
    classification_losses = []
    top1_accuracies = []
    top5_accuracies = []

    best_top1 = 0.0  # Initialize best top1 accuracy

    for epoch in range(epochs):
        epoch_losses = []
        top1 = []
        top5 = []

        for images, targets in train_loader:
            images = images.to(device)
            targets = targets.to(device)

            # Extract features without gradient computation
            with torch.no_grad():
                _, features = model(images)

            # Train classifier on frozen features
            logits = classifier(features)
            loss = F.cross_entropy(logits, targets)

            optimizer_classifier.zero_grad()
            loss.backward()
            optimizer_classifier.step()

            # Calculate accuracies (assuming accuracy function is defined)
            acc1, acc5 = accuracy(logits, targets, topk=(1, 5))
            epoch_losses.append(loss.item())
            top1.append(acc1.item())
            top5.append(acc5.item())

        avg_loss = sum(epoch_losses) / len(epoch_losses)
        avg_top1 = sum(top1) / len(top1)
        avg_top5 = sum(top5) / len(top5)
        print(f'Classifier Training - Epoch {epoch}: Loss = {avg_loss:.4f}, '
              f'Top-1 Acc = {avg_top1:.2f}%, Top-5 Acc = {avg_top5:.2f}%')

        classification_losses.append(avg_loss)
        top1_accuracies.append(avg_top1)
        top5_accuracies.append(avg_top5)

        # Save the classifier checkpoint if the top1 accuracy has improved
        if avg_top1 > best_top1:
            best_top1 = avg_top1
            torch.save(classifier.state_dict(), "classifier_best.pth")
            print(f"Improved top-1 accuracy detected. Saving classifier checkpoint at epoch {epoch}.")

    return classification_losses, top1_accuracies, top5_accuracies


# Example usage of separate training
def train_separately(train_loader, model, classifier, device, maxepochs=2):
    # Stage 1: Train rotation prediction
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
    rotation_losses, rotation_accuracies = train_rotation_model(
        train_loader, model, optimizer, maxepochs, device
    )

    # Stage 2: Train classifier
    optimizer_classifier = torch.optim.Adam(classifier.parameters(), lr=0.01)
    class_losses, top1_acc, top5_acc = train_classifier(
        train_loader, model, classifier, optimizer_classifier, maxepochs, device
    )

    return {
        'rotation': {'losses': rotation_losses, 'accuracies': rotation_accuracies},
        'classification': {'losses': class_losses, 'top1_acc': top1_acc, 'top5_acc': top5_acc}
    }
```
